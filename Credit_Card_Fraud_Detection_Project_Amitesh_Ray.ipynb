Importing the Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading CSV to pandas dataframe
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset informations
credit_card_data.info()
```

```python
#check missing values
credit_card_data.isnull().sum()
```

```python
#check distribution of legit transactions and fraudulent transactions
#0 is normal
#1 is fraud
credit_card_data['Class'].value_counts()
```

Highly unbalanced dataset

```python
#separating data for analysis
legit=credit_card_data[credit_card_data.Class == 0]
fraud=credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#COMPARISION
credit_card_data.groupby('Class').mean()
```

```python
#Undersampling
#Build a Sample Data set from original, having similar distribution of legit/fraud transaction
#fraud count=93
legit_sample=legit.sample(n=93)
```

```python
#concat
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into Features and Targets
X= new_dataset.drop(columns='Class',axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting Data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training
#Logistic Regression, as it is binary classification
model=LogisticRegression()
```

```python
#training the model with Training Data
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
#Accuracy Score
#Train Data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy: ',training_data_accuracy)
```

Accuracy is 98.64% in train data

```python
#Test Data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy: ',test_data_accuracy)
```

Accuracy is 89.47% for test data

There is no Overfitting or Underfitting problem, as test and training data accuracy values are quite close
